```python
# Importation des librairies Python nécessaires à la résolution du problème
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score
```

```python
# Chargement des données du Breast Cancer Wisconsin (diagnostic) Dataset
data = pd.read_csv('data.csv')
```

```python
# Nettoyage
data = data.drop(columns='Unnamed: 32')
```

```python
# Normalisation des étiquettes 'diagnosis' avec LabelEncoder
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])
```

```python
# Séparation des caractéristiques et de l'étiquette
features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean',
            'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean', 'radius_se',
            'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se',
            'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst',
            'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst',
            'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']

X = data[features]
y = data['diagnosis']

# Split jeu de test, jeu d'entraînement
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

```python
# Scaling des données
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
# Analyse univariée
for feature in features:
    # Afficher des statistiques descriptives pour chaque variable
    print(f"\nStatistiques descriptives pour {feature}:\n{data[feature].describe()}")

    # Afficher un histogramme pour chaque variable
    plt.figure(figsize=(8, 5))
    sns.histplot(data[feature], kde=True)
    plt.title(f"Histogramme pour {feature}")
    plt.xlabel(feature)
    plt.ylabel("Fréquence")
    plt.show()
```

```python
# Initialisation du modèle de régression logistique
model = LogisticRegression(C=0.1, max_iter=10000)

# Entraînement du modèle sur les données d'entraînement
model.fit(scaled_X_train, y_train)

# Prédiction sur le jeu d'entraînement
y_pred_train = model.predict(scaled_X_train)

# Prédiction sur le jeu de test
y_pred_test = model.predict(scaled_X_test)
```

```python
# Obtenir les coefficients de régression logistique
coefficients = model.coef_[0]

# Créer un DataFrame pour afficher les coefficients par variable
coefficients_df = pd.DataFrame({'Variable': features, 'Coefficient': coefficients})

# Trier le DataFrame par valeur absolue des coefficients pour visualiser l'impact
coefficients_df['AbsoluteCoefficient'] = coefficients_df['Coefficient'].abs()
coefficients_df = coefficients_df.sort_values(by='AbsoluteCoefficient', ascending=False)

# Afficher les coefficients par variable
print("Coefficients de Régression Logistique:")
print(coefficients_df)

# Afficher un graphique des coefficients
plt.figure(figsize=(12, 8))
sns.barplot(x='Coefficient', y='Variable', data=coefficients_df, palette='viridis')
plt.title('Importance des Variables dans la Détection du Cancer du Sein')
plt.xlabel('Coefficient de Régression Logistique')
plt.ylabel('Variable')
plt.show()
```

```python
# Calcul de la matrice de confusion pour le jeu d'entraînement
cm_train = confusion_matrix(y_train, y_pred_train)
print("Matrice de Confusion pour le jeu d'entraînement :")
print(cm_train)
```

```python
# Calcul de la matrice de confusion pour le jeu de test
cm_test = confusion_matrix(y_test, y_pred_test)
print("Matrice de Confusion pour le jeu de test :")
print(cm_test)
```

```python
# Affichage de la classification_report pour le jeu d'entraînement
print("Classification Report pour le jeu d'entraînement:")
print(classification_report(y_train, y_pred_train))
```

```python
# Affichage de la classification_report pour le jeu de test
print("Classification Report pour le jeu de test:")
print(classification_report(y_test, y_pred_test))
```

```python
# Calcul des scores de probabilité prédits par le modèle pour la classe positive
y_score_train = model.predict_proba(scaled_X_train)[:, 1]
y_score_test = model.predict_proba(scaled_X_test)[:, 1]

# Calcul des taux de faux positifs (FPR), des taux de vrais positifs (TPR) et des seuils pour la courbe ROC
fpr_train, tpr_train, _ = roc_curve(y_train, y_score_train)
fpr_test, tpr_test, _ = roc_curve(y_test, y_score_test)

# Calcul de l'aire sous la courbe ROC (ROC AUC) pour évaluer la performance du modèle
roc_auc_train = roc_auc_score(y_train, y_score_train)
roc_auc_test = roc_auc_score(y_test, y_score_test)
```

```python
# Affichage de l'aire sous la courbe ROC pour le jeu d'entraînement
print('ROC AUC Score pour le jeu d\'entraînement:', roc_auc_train)
```

```python
# Affichage de l'aire sous la courbe ROC pour le jeu de test
print('ROC AUC Score pour le jeu de test:', roc_auc_test)
```

```python
# Affichage de la courbe ROC pour le jeu d'entraînement
plt.figure(figsize=(8, 6))
plt.plot(fpr_train, tpr_train, label=f'ROC Curve - Train (AUC = {roc_auc_train:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='grey', label='Random')
plt.title('Receiver Operating Characteristic - Train')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
# Affichage de la courbe ROC pour le jeu de test
plt.figure(figsize=(8, 6))
plt.plot(fpr_test, tpr_test, label=f'ROC Curve - Test (AUC = {roc_auc_test:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='grey', label='Random')
plt.title('Receiver Operating Characteristic - Test')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
